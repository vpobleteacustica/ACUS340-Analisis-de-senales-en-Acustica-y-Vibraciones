La **energía** ([Wikipedia](https://en.wikipedia.org/wiki/Energy_(signal_processing%29); FMP, p. 66) de una señal corresponde a la magnitud total de la señal. Para señales de audio corresponde aproximadamente a qué tan fuerte es la señal. La energía en una señal se define como:

$$ \sum_n \left| x(n) \right|^2 $$

La **energía cuadrática media (RMSE)** en una señal se define como

$$ \sqrt{ \frac{1}{N} \sum_n \left| x(n) \right|^2 } $$
